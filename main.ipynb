```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fcfs = pd.read_csv('fcfs.csv')
fcfs_delayed = pd.read_csv('fcfs_delayed.csv')
rr = pd.read_csv('rr.csv')
time_series = pd.Series(np.linspace(0, 700, 10))
```

### Generate separate graph

```python
def plot_csv_data(df, label_prefix):
    plt.figure(figsize=(10, 6))
    plt.plot(df['process'], df['avg_time'], marker='o', label=f'{label_prefix} avg_time')
    plt.plot(df['process'], df['elapsed_time'], marker='o', label=f'{label_prefix} elapsed_time')
    plt.plot(df['process'], df['avg_wait_time'], marker='o', label=f'{label_prefix} avg_wait_time')
    plt.plot(df['process'], df['burst_time'], marker='o', label=f'{label_prefix} burst_time')
    plt.xlabel('Processes')
    plt.ylabel('Time (in secs)')
    plt.title(f'{label_prefix} Scheduling Algorithm')
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=1)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot each DataFrame separately
plot_csv_data(fcfs, 'FCFS')
plot_csv_data(fcfs_delayed, 'FCFS Delayed')
plot_csv_data(rr, 'RR')
```

### Generate summary graph

```python
def plot_csv_data(df, label_prefix, ax):
    ax.plot(df['process'], df['avg_time'], marker='o', label=f'{label_prefix} avg_time')
    ax.plot(df['process'], df['elapsed_time'], marker='o', label=f'{label_prefix} elapsed_time')
    ax.plot(df['process'], df['avg_wait_time'], marker='o', label=f'{label_prefix} avg_wait_time')
    ax.plot(df['process'], df['burst_time'], marker='o', label=f'{label_prefix} burst_time')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

plot_csv_data(fcfs, 'FCFS', axs[0, 0])
axs[0, 0].set_title('FCFS')
axs[0, 0].legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=1)
axs[0, 0].grid(True)

plot_csv_data(fcfs_delayed, 'FCFS Delayed', axs[0, 1])
axs[0, 1].set_title('FCFS Delayed')
axs[0, 1].legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=1)
axs[0, 1].grid(True)

plot_csv_data(rr, 'RR', axs[1, 0])
axs[1, 0].set_title('RR')
axs[1, 0].legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=1)
axs[1, 0].grid(True)

axs[1, 1].axis('off')

for ax in axs.flat:
    ax.set(xlabel='Processes', ylabel='Time (in secs)')

fig.suptitle('Comparison of Scheduling Algorithms', fontsize=16, y=1.02)
# plt.subplots_adjust(top=2)

plt.tight_layout()
plt.show()
```
